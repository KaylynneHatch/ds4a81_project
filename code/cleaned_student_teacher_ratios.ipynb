```python
#All but the last dataset has an extra column (the last is missing one) - Luckily, we really only need 3 columns: State, Year, and Pupil/Teacher Ratio. We'll get rid of: Total Students and Full-Time Equivalent (FTE) Teachers
sute910 = sute910.rename(columns={sute910.columns[3]: "Ratio" })
Sub_2009 = sute910[['State Name', 'Ratio', 'Year']]
sute1011 = sute1011.rename(columns={sute1011.columns[3]: "Ratio" })
Sub_2010 = sute1011[['State Name', 'Ratio', 'Year']]
sute1112 = sute1112.rename(columns={sute1112.columns[3]: "Ratio" })
Sub_2011 = sute1112[['State Name', 'Ratio', 'Year']]
sute1213 = sute1213.rename(columns={sute1213.columns[3]: "Ratio" })
Sub_2012 = sute1213[['State Name', 'Ratio', 'Year']]
sute1314 = sute1314.rename(columns={sute1314.columns[3]: "Ratio" })
Sub_2013 = sute1314[['State Name', 'Ratio', 'Year']]
sute1415 = sute1415.rename(columns={sute1415.columns[3]: "Ratio" })
Sub_2014 = sute1415[['State Name', 'Ratio', 'Year']]
sute1516 = sute1516.rename(columns={sute1516.columns[3]: "Ratio" })
Sub_2015 = sute1516[['State Name', 'Ratio', 'Year']]
sute1617 = sute1617.rename(columns={sute1617.columns[3]: "Ratio" })
Sub_2016 = sute1617[['State Name', 'Ratio', 'Year']]
sute1718 = sute1718.rename(columns={sute1718.columns[3]: "Ratio" })
Sub_2017 = sute1718[['State Name', 'Ratio', 'Year']]
sute1819 = sute1819.rename(columns={sute1819.columns[2]: "Ratio" })
Sub_2018 = sute1819[['State Name', 'Ratio', 'Year']]
```

```python
#Combine all dataframes
subsets = [Sub_2009,Sub_2010,Sub_2011,Sub_2012,Sub_2013,Sub_2014,Sub_2015,Sub_2016,Sub_2017,Sub_2018]
compiled_stuteach = pd.concat(subsets)
```

```python
#Rename 'State Name' column to just 'State'
compiled_stuteach = compiled_stuteach.rename(columns = {'State Name':'State'})
```

```python
#Remove any rows that are territories (and not states)
compiled_stuteach = compiled_stuteach[~compiled_stuteach['State'].str.contains('BUREAU OF INDIAN EDUCATION')]
compiled_stuteach = compiled_stuteach[~compiled_stuteach['State'].str.contains('AMERICAN SAMOA')]
compiled_stuteach = compiled_stuteach[~compiled_stuteach['State'].str.contains('DEPARTMENT OF DEFENSE EDUCATION ACTIVITY')]
compiled_stuteach = compiled_stuteach[~compiled_stuteach['State'].str.contains('GUAM')]
compiled_stuteach = compiled_stuteach[~compiled_stuteach['State'].str.contains('U.S. VIRGIN ISLANDS')]
compiled_stuteach = compiled_stuteach[~compiled_stuteach['State'].str.contains('NORTHERN MARIANAS')]
compiled_stuteach = compiled_stuteach[~compiled_stuteach['State'].str.contains('PUERTO RICO')]
```

```python
#Convert all State names from uppercase to camel case
compiled_stuteach['State'] = compiled_stuteach['State'].str.title()
```
